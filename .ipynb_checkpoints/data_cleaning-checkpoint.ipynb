```python
mapp = {val:key for key, val in pmap.items()}
mapp1 = {val:key for key, val in pmap.items()}
mapp['D.J. Moore'] = 3915416
mapp['D.J. Chark'] = 3115394
```

```python
proj = proj.dropna(subset=['adp'])
```

```python
proj['eid1'] = proj['player'].map(mapp)
```

```python
mask = proj['eid1'].isna()
```

```python
fuzz_names = proj[mask].copy()
```

```python
fuzz_names['fuzz_match'] = fuzz_names['player'].apply(lambda x: process.extractOne(x, list(mapp.keys()))[0])
```

```python
proj['espn_names'] = proj['player'].map(fuzz_names.set_index('player')['fuzz_match'])
```

```python
proj['en1'] = proj['eid1'].map(pmap)
```

```python
proj['en1'] = np.where(proj['en1'].isna(), proj['espn_names'], proj['en1'])
```

```python
proj['espn_id'] = proj['en1'].map(mapp1)
```
