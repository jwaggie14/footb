```python
a = pd.read_csv(file(2017))
b = pd.read_csv(file(2018))
c = pd.read_csv(file(2019))
```

```python
hist = a.append(b).append(c)
```

```python
hist['rookie'] = hist['exp'] == 0
```

```python
hist['position'].unique()
```

```python
hist = hist[hist['position'].isin(['RB','QB','WR','TE'])]
```

```python
regressions = {}
```

```python
import statsmodels.api as sm
```

```python
positions = ['RB','QB','WR','TE']
```

```python
hist.columns
```

```python
hf = hist[hist['adp'] <= 140]
```

```python
import matplotlib.pyplot as plt
```

```python
color = ['red','green','blue','orange']

fig, ax = plt.subplots(figsize=(10,5))
for i in range(4):
    hf[hf['position'] == positions[i]].plot.scatter(x='adp',y='points', color=color[i], ax=ax)
```

```python
color = ['red','green','blue','orange']

fig, ax = plt.subplots(figsize=(10,5))
for i in range(4):
    hf[hf['position'] == positions[i]].plot.scatter(x='adp',y='actualPoints', color=color[i], ax=ax)
```

```python
regressions = {}
reg2 = {}
reg3 = {}
reg4 = {}
for i in range(4):
    data = hf[hf['position'] == positions[i]][['adp','actualPoints','points']].dropna()
    X = sm.add_constant(data['adp'])
    y = data['actualPoints']
    regressions[i] = sm.OLS(y,X).fit()
    X2 = sm.add_constant(data['points'])
    reg2[i] = sm.OLS(y,X2).fit()
    X3 = sm.add_constant(data[['adp','points']])
    reg3[i] = sm.OLS(y,X3).fit()
    reg4[i] = sm.RLM(y,X2, M=sm.robust.norms.HuberT()).fit()
```
